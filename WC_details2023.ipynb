```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.espncricinfo.com/records/tournament/team-match-results/icc-cricket-world-cup-2023-24-15338"
main_page = requests.get(url)
print (main_page.status_code) #Checking for menctioned link valid or not
```

```python
#Extracting text from menctioned page
soup = BeautifulSoup(main_page.text,'html.parser')
```

```python
#Extracting Headings
hadders = soup.find('thead',class_ = "ds-bg-fill-content-alternate ds-text-left").find_all('tr')
for title in hadders:
    row = title.find_all('td',class_ = "ds-min-w-max ds-text-right")
    new_rows = [data.text.strip() for data in title]
    print (new_rows)
```

```python
#Creating DataFrame with hadders
data_frame = pd.DataFrame(columns = new_rows)
```

```python
data_frame
```

```python
#Extracting Data from tags
matches =soup.find('tbody').find_all('tr')
for item in matches:
    match= item.find_all('td',class_ = "ds-min-w-max ds-text-right")
    indv_match = [data.text.strip() for data in item]
    length = len(data_frame)
    data_frame.loc[length] = indv_match
```
